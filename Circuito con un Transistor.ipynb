# Circuito con un Transistor

## Introducción

Uno de los circuitos con un transistor más comunes es el que muestra en la siguiente figura
![CircuitoBasico](images/parametros_hibridos - Original.png)
el circuito se espera que cumpla con la relación
\begin{equation}
V_{R_{L}}=A_{V}V_{i},
\end{equation}
donde $A_{V}$ es una constante que se conoce como ganancia de voltaje y $V_{R_{L}}$ es el voltaje en la resistencia de carga. Se tiene que si $A_{v}>1$ se dice que el circuito es un amplificador, si $A_{V}<1$ se conoce como atenuador.

Para determinar el comportamiento del circuito es necesario obtener su modelo mátematico, dado que las ecuaciones pueden llegar a ser complejas al considerar las ecuaciones de estado solidos correspondientes a la dinámica del transistor, en general, se realiza un analisis aproximado.


## Analisís Aproximado

Para poder simplificar el analisís del circuito lo primero que se considera es que:

\begin{equation}\label{eq:ca}
C_{A}\rightarrow\infty.
\end{equation}

Con la condición \eqref{eq:ca} se tiene que:
\begin{equation}
X_{C}=
\begin{cases}
0 & \omega=0 \\
\infty & \omega > 0
\end{cases}
\end{equation}

entonces se puede proponer que el circuito se puede separar  en dos circuitos, en función de las fuentes de DC y AC como se muestra en la siguiente figura. El primero (a) corresponde al efecto de la fuente de DC $V_{cc}$ y el segundo circuito (b) corresponde a la fuente de AC $V_{i}$.

![Separación según la fuente](images/parametros_hibridos - separacion.png "Separación según la fuente")

 

### Analisís en DC

![BJT en DC](images/bjt_dc2.png)
de este circuito se obtienen para la maya $\color{blue}{1}$ la expresión \eqref{eq:cir1may1}:

\begin{equation}\label{eq:cir1may1}
\color{blue}{V_{cc}=V_{R_{c}}+V_{ce}}
\end{equation}
del maya $\color{red}{2}$ se obtiene la expresión \eqref{eq:cir1may2}
\begin{equation}\label{eq:cir1may2}
\color{red}{V_{cc}=V_{R_{b}}+V_{be}}
\end{equation}

Si suponemos que el transistor se encuentra en la region lineal entonces se tiene que:


$$
\begin{align}
\label{eq:ic} i_{c} &= \beta i_{b}  \\
\label{eq:Ie} i_{e} & = i_{c} + i_{b} \\
\label{eq:icIe} i_{e} & = \alpha i_{c}   \\ 
\label{eq:alhpa} \alpha & = \frac{\beta}{\beta+1} 
\end{align}
$$


Considerando la **ley de Ohm** en la expresión \eqref{eq:cir1may1} se tiene que:

\begin{equation}\label{eq:cir1may1b}
\color{blue}{V_{cc}=i_{R_{c}}R_{c}+V_{ce}}
\end{equation}

dado que la resistencia $R_{c}$ esta en serie con el transitor, entonces se puede considerar que $\color{green}{i_{c}=i_{R_{c}}}$, por lo que la expresión \eqref{eq:cir1may1b} se puede reescribir como:

\begin{equation}\label{eq:cir1may1c}
\color{blue}{V_{cc}=i_{c}R_{c}+V_{ce}}
\end{equation}

En la expresión \eqref{eq:cir1may1c} se tiene que $i_{c}$ y $V_{ce}$ son parámetros del transistor y por lo tanto se consideran incognitas, lo cual lleva a que no se pude encontrar una solución única. Utilizando un analisís similar en la expresión \eqref{eq:cir1may2} se tiene que:

\begin{equation}\label{eq:cir1may2a}
\color{red}{V_{cc}=i_{b}R_{b}+V_{be}}
\end{equation}

En la expresión \eqref{eq:cir1may2a} se tiene que idelamente en un transistor de silicio $V_{be}=0.7$, por lo que se tiene que:

\begin{equation}\label{eq:cir1may2b}
\color{red}{i_{b}=\frac{V_{cc}+V_{be}}{R_{b}}}
\end{equation}

De la expreesión \eqref{eq:cir1may2b} se observa que el valor de la corriente $i_{b}$ es determinado por elementos externos al transistor ($V_{cc}$ y $R_{b}$). Además, si se considera la expresión \eqref{eq:ic} en \eqref{eq:cir1may1c} entonces se tiene que el valor de $i_{c}$ y de $V_{ce}$ se determinán por los elementos externos al tranistor. 

Si los elemento externos determinan los posibles valores de los parámetros del transistor ($i_{b}$,$i_{c}$ y $V_{ce}$) entonces, sin importar el valor de $\beta$ se puede representar graficamente cualquier posible valor de $i_{c}$ y $V_{ce}$ utilizando la expresión \eqref{eq:cir1may1c}, se tiene que el máximo valor de $i_{c}$ se obtiene cuando $V_{cc}=0$ y su expresión es:

\begin{equation}\label{eq:cir1icmax}
\color{blue}{\text{max}(i_{c})=\frac{V_{cc}}{R_{c}}}
\end{equation}

el valor máximo de $V_{ce}$ se obitene cuando $i_{c}=0$ y su expresión es:

\begin{equation}\label{eq:cir1vccmax}
\color{blue}{\text{max}(V_{ce})=V_{cc}}
\end{equation}

De las expresiones \eqref{eq:cir1may1c},\eqref{eq:cir1icmax} y \eqref{eq:cir1vccmax} se puede concluir que un incremento en $i_{c}$ o $V_{ce}$ involucra un decremento en  $V_{ce}$ o $i_{c}$ respectivemente.  Dado que el transitor trabaja en el primer cuadrante de la grafica IV, entonces todo posible valor de $i_{c}$ y $V_{ce}$ se puede representar a través de la siguiente gráfica

<p align="center"><img src="/files/images/recta_carga.png" alt="una image" title="Recta de Carga" style="max-width:50%" ></p>

La recta en la figura anterior es conocida como **Recta de Carga**, la cual se puede definir como: **La representación gráfica de todos los posibles valores de $i_{c}$ y $V_{ce}$**. La Recta de carga, junto con las curvas caracteristicas del transistor nos permiten determinar el comportamiento del circuito, al poder determinar si en que región de funcionamiento se encuentra el transistor, es importante recordar una de las hipotesis en el analisís es considerar que el transistor se encuentra en la región lineal.

<p align="center"><img src="/files/images/recta_carga3.png" alt="una image" title="Recta de Carga" style="max-width:50%" ></p>

Aunque la Recta de Carga representa todos los posibles valores de $i_{c}$ y $V_{ce}$, en la práctica se tiene que $i_{c}$ y $V_{ce}$ tienen un valor único,  a ese valor especifico de $i_{c}$ y $V_{ce}$ debido a los valores de las fuentes en el  circuito (DC y AC) se le llama **Punto de Operación**.

<p align="center"><img src="/files/images/recta_carga4.png" alt="una image" title="Recta de Carga" style="max-width:50%" ></p>

Dadas las expresiones \eqref{eq:ic} y \eqref{eq:cir1may1c} se puede considerar que  $V_{ce}$ es una señal de AC con un componente de DC (una señal de AC con un *offset*).  El punto de operación cuando la señal de AC es nula corresponde al compente de DC, y se llama **punto de reposo** es denota con la letra $Q$.  En la siguiente figura se presenta un bosquejo de la evolución en tiempo de $V_{ce}$. 

<p align="center"><img src="/files/images/vce1.png" alt="una image" title="Recta de Carga" style="max-width:50%" ></p>

La amplitud de la señal $V_{ce}$ está limitado por la expresión \eqref{eq:cir1vccmax} y la naturaleza del transitor que requiere que $V_{ce}\geq0$ por lo que se tiene que:

\begin{equation}\label{eq:vceac}
0\leq V_{ce}(AC) \leq V_{cc}
\end{equation}


Dado que el comportamiento de $V_{ce}$ esta relacionado con $i_{c}$ por la expresión eqref{eq:cir1may1c}, entoces se concluye que $i_{c}$ presente un comporamiento similar a $V_{ce}$ pero de forma inversa, cuando $V_{ce}$ aumenta $i_{c}$ disminuye. Si consideramos el comportamiento de AC sobre la recta de carga, se tendra que el punto de reposo es el origen del sistema de AC. Además que los limites del comportamiento de AC por la expresión \eqref{eq:vceac} y de forma similar $i_{c}$ por:

\begin{equation}\label{eq:icac}
0\leq i_{c}(AC) \leq \frac{V_{cc}}{R_{c}}
\end{equation}

Si se considera que la señal de AC en el circuito es simetrica

$$
\text{max}(V_i(t))=-\text{min}(V_i(t))
$$

entonces se tiene que el mejor valor para el punto de reposos $Q$ es a la mitad de la recta de carga
\begin{eqnarray}
i_{c}(Q)& = &\frac{\text{max}(i_{c})}{2}, \label{eq:icq}\\
V_{ce}(Q)& = &\frac{\text{max}(V_{ce})}{2}, \label{eq:vceq}
\end{eqnarray}
graficamente se representa en la siguiente figura

<p align="center"><img src="/files/images/recta_ac1.png" alt="una image" title="Recta de Carga" style="max-width:50%" ></p>

De las expresiones \eqref{eq:icac} y \eqref{eq:vceac} se puede concluir que el diseño del citcuito en DC debe considerar las necesidades en la salida del circuito en AC. De no **cumplir** con las  expresiones \eqref{eq:icac} y \eqref{eq:vceac} el circuito seguira funcionando correctamente en DC, lo que ocurrira es que no se tendrá la salida en AC deseada

## Analisis en AC

Para poder realizar el analisis en AC es necesario considerar que el mismo ya esta polarizado y en el punto de operación $Q$ está en la región lineal. 

El primer paso es realizar la separación de fuentes y dejar únicamente la fuente de AC en el circuito el cual se muestra en la siguiente figura

![Circuito en AC](images/parametros_hibridos - AC.png "Circuito en AC")

Dada la dificultad de analizar el circuito con las ecuaciones de estado solido que gobiernan la naturaleza del tranistos, se propone un anlaisis aproximado utilizando la **tecnica de bipuertos** usada en circuitos electricos. En el caso del transistor BJT se utilizan los parámetros *H*, esta tecnica de analisis se basa en el que cualquier circuito electrico por complejo que sea, si se le pueden asociar 2 puertos se puede representar de forma más simple. 

![Caja Negra](images/parametros_hibridos - caja negra.png "Caja Negra")

El concepto básico, es que a cada puerto se le asigna un voltaje $V_{x}$ y una corriente $i_{x}$. El siguiente es considerar alguno de los dos como independiente y al otro como dependiente, es decir, que al aplicar uno de los dos la otra variable es consecuencia. De está forma si se aplica voltaje en el puerto 1 $V_{1}$ se espera que se produsca la corriente $i_{1}$. Además, la parte importante es que trata de mostrar como la energía inyectada en el puerto 1 es afecta al puerto 2 y cuando se aplica energía en el puerto 2 como afecta al puerto 1. 

Existen varias formas de representar los bipuertos, estas diferencias surgen de que se inyecta (corriente o voltaje) en cada puerto. En el caso del transitor existen diversas configuraciones posibles para usar los parámetros hibridos, la que se muestra en la siguente figura se conoce como parámetros hibridos de emisor común. Básicamente para que se puede usar está configuración es necesario que la energía se inyecte por la base y el colector, mientras que el emisor se conecta a tierra (Puede ser a través de una red electrica).

![Parámetros Híbridos del transisotr BJT emisor común](images/parametros_hibridos_caja.png "Parámetros Híbridos del transisotr BJT emisor común")

\begin{eqnarray}
V_{1}&=&h_{11}i_{1}+h_{12}V_{2} \\
i_{2}&=&h_{21}i_{1}+h_{22}V_{2}
\end{eqnarray}

### Analisis de DC

![Seguidor de Emisor Thevenin](images/seguidor de voltaje - thevnin.png "Seguidor de Emisor Thevenin")
Se tiene que el circuito en DC tiene la ecuaciones

\begin{eqnarray}
V_{cc}&=&V_{ce}+V_{R_{E}} \label{eq:seguidor1}\\
V_{th}&=&V_{R_{th}}+V_{BE}+V_{R_{E}} \label{eq:seguidor2}
\end{eqnarray}

Si consideramos \eqref{eq:seguidor1}, se tiene que los puntos de la recta de carga
\begin{equation}
\text{max}(V_{CE})=V_{cc}
\end{equation}
y
\begin{equation}
\text{max}(i_{c})=\frac{V_{cc}}{R_{E}/\alpha}
\end{equation}

si proponemos que se tenga máxima simetría

\begin{eqnarray}
V_{CE}=\frac{V_{cc}}{2} \\
i_{c}=\frac{V_{cc}}{2R_{E}/\alpha}
\end{eqnarray}

la parte de $i_{b}$ se diseña utilizando el teorema de thevenin, donde se tiene que

\begin{eqnarray}
R_{1}&=&\frac{V_{cc}}{V_{th}}R_{th} \label{eq:seguidorr1} \\
R_{2}&=&\frac{R_{th}V_{cc}}{V_{cc}-V_{th}} \label{eq:seguidorr2}
\end{eqnarray}

### Analisis de AC

![Parámetros Hibridos Seguidor](images/seguidor_puertos.png "Parámetros Hibridos Seguidor")
Se tiene que el voltaje de salida $V_{R_{L}}$ está en paralelo con $V_{R_{E}}$, y se tiene que 

$$
V^{*}=V_{h_{ie}}+V_{R_{L}}
$$

y se tiene que

$$
V^{*}=\frac{R^{*}}{R^{*}+R_{i}}V_{i}
$$

donde

$$
R^{*}=R_{th}|(h_{ie}+(\beta+1)(R_{E}|R_{L}))
$$

sustituyendo se tiene que 

$$
V_{h_{ie}}+V_{R_{L}}=\frac{R^{*}}{R^{*}+R_{i}}V_{i}
$$

se tiene que

$$
V_{h_{ie}}=\frac{V^{*}}{h_{ie}+(\beta+1)(R_{E}|R_{L})}h_{ie}
$$

entonces

\begin{align*}
V_{R_{L}}&=\frac{R^{*}}{R^{*}+R_{i}}V_{i}-\frac{\frac{R^{*}}{R^{*}+R_{i}}h_{ie}}{h_{ie}+(\beta+1)(R_{E}|R_{L})}V_{i} \\
&=\frac{R^{*}}{R^{*}+R_{i}}\left ( 1-\frac{h_{ie}}{h_{ie}+(\beta+1)(R_{E}|R_{L})}\right )V_{i}
\end{align*}

si se cumple que $R^{*} \gg R_{i}$ y $(\beta+1)(R_{E}|R_{L}) \gg h_{ie}$ la ganacia se puede definir como:

$$
A_{V}=\frac{R^{*}}{R^{*}+R_{i}}\left ( 1-\frac{h_{ie}}{h_{ie}+(\beta+1)(R_{E}|R_{L})}\right )\approx 1
$$

En el caso de la corriente se tiene que:

$$
i_{b}=\left(1-\frac{h_{ie}+(\beta+1)(R_{E}|R_{L})}{h_{ie}+(\beta+1)(R_{E}|R_{L})+R_{th}}\right)i_{i}
$$

mientras que $i_{R_{L}}$ es dado por

$$
i_{R_{L}}=\left(1-\frac{R_{L}}{R_{L}+R_{E}} \right)i_{C}
$$

considerando que $i_{C}=\beta i_{b}$, se tiene que la ganacia de corrientes es:

$$
A_{i}=\beta \left(1-\frac{h_{ie}+(\beta+1)(R_{E}|R_{L})}{h_{ie}+(\beta+1)(R_{E}|R_{L})+R_{th}}\right)\left(1-\frac{R_{L}}{R_{L}+R_{E}} \right)
$$

si se cumple que $R_{th} \gg h_{ie}+(\beta+1)(R_{E}|R_{L})$ entonces
$$
A_{i}\approx\beta \left(1-\frac{R_{L}}{R_{L}+R_{E}} \right)
$$

```python
filename = 'seguidor.txt'

# Using the newer with construct to close the file automatically.
with open(filename) as f:
    data2 = f.readlines()
k=0
inde=[]
for i in data2:
    if 'Step' in i:
        inde=append(inde,k)
    k +=1
inde=append(inde,len(data2))
k=len(inde)

leninde=[]
for i in range(0,k-1):
    leninde=append(leninde,inde[i+1]-inde[i])
tam1=len(inde)
matplotlib.rcParams.update({'font.size': 20,'text.usetex': True})
plt.figure(figsize=(14,10), dpi=150)


for i in range(tam1-1):
    time=np.array([],dtype=float32)
    Vi=np.array([],dtype=complex64)
    for k in range(int(inde[i])+1,int(inde[i+1])):
        temp=data2[k].strip().split('\t')
        time=np.append(time,temp[0])
        tempcom=temp[1].split(',')
        Vi=np.append(Vi,float(tempcom[0])+1j*float(tempcom[1]))
    plt.plot(time,10*np.log(np.abs(Vi)),linewidth=3)
ax=plt.gca()
ax.set_xscale('log')
plt.grid(b=True, which='major', color='k', linestyle='-.',linewidth=2)
plt.grid(b=True, which='minor', color='k', linestyle='--',linewidth=1)
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud')
plt.legend(['0.1$\mu$','1$\mu$','10$\mu$','100$\mu$','1000$\mu$'],loc=4)
plt.show()
```
