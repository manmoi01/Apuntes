```python
import sqlite3 as sq
import numpy as np
from datetime import datetime
import matplotlib.pylab as plt
import matplotlib.dates as dts
import matplotlib

matplotlib.rcParams.update({'font.size': 16})

class DataCursor(object):
    text_template = 'pH: %0.2f\nFecha: %s'
    x, y = 0.0, 0.0
    xoffset, yoffset = -20, 20
    text_template = 'pH: %0.2f\nFecha: %s'

    def __init__(self, ax):
        self.ax = ax
        self.annotation = ax.annotate(self.text_template, 
                xy=(self.x, self.y), xytext=(self.xoffset, self.yoffset), 
                textcoords='offset points', ha='right', va='bottom',
                bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0')
                )
        self.annotation.set_visible(False)

    def __call__(self, event):
        self.event = event
        # xdata, ydata = event.artist.get_data()
        # self.x, self.y = xdata[event.ind], ydata[event.ind]
        self.x, self.y = event.mouseevent.xdata, event.mouseevent.ydata
        if self.x is not None:
            self.annotation.xy = self.x, self.y
            self.annotation.set_text(self.text_template % (self.y,dts.num2date(self.x).strftime('%d/%m %H:%M')))
            self.annotation.set_visible(True)
            event.canvas.draw()

connection = sq.connect('datos1.db')
cursor = connection.cursor()
k=450
k2=5
s='select * from datos order by id desc limit '+str(k)
cursor.execute(s)
record=cursor.fetchall()
ph=np.zeros(k)
id=np.zeros(k)
dates=np.array(np.zeros(k),dtype=object)
for i in range(0,k):
    ph[i]=record[i][2]
    id[i]=record[i][0]
    dates[i]=datetime.strptime(record[i][1], "%Y-%m-%d %H:%M")

fig = plt.figure(1, figsize=(20, 8))
ax=fig.gca()
line, = ax.plot(dates[k2:k-1],ph[k2:k-1],linewidth=2.0)


#ax.xaxis.set_major_locator(matplotlib.dates.WeekdayLocator(byweekday=matplotlib.dates.MO))

ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%d/%m\n%H:%M'))
ax.grid(True)
plt.title('Evolución en Tiempo del pH')
plt.xlabel('Tiempo')
plt.ylabel('pH')
plt.ylim([0,14])
plt.subplots_adjust(left=0.05, right=0.97, top=0.94, bottom=0.12)
fig.canvas.mpl_connect('pick_event', DataCursor(plt.gca()))
line.set_picker(5) # Tolerance in points
plt.show()
```
