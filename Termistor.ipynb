#Sistema usando un termistor

Un termistor es un elemento transductor que cambia su impedancia en función de la temperatura, su comportamiento es descrito por la ecuación de Stainhart-Hart

\begin{equation}\label{eq:stainharthart}
\frac{1}{T}=A+B\ln(R)+C\ln(R)^{3}
\end{equation}

donde $T$ es la temepratura en grandos kelvin (ºk). Los parametros $A,B$y $C$ varían de acuerdo al termistor, por lo que es necesario calcularlo para cada uno. Su simbolo es

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import csv
#rc('text', usetex=True)
```

```python
def getColumn(filename, column):
    results = csv.reader(open(filename), delimiter=",")
    return [result[column] for result in results]

time = getColumn("TS91ResTemp.csv",0)
volt = getColumn("TS91ResTemp.csv",1)

# Create a new figure of size 14x10 points, using 150 dots per inch
matplotlib.rcParams.update({'font.size': 18})
plt.figure(figsize=(14,10), dpi=150)

plt.xlabel("Temperatura (ºC)")
plt.ylabel("Resistencia $\Omega$")
plt.grid('on')
plt.plot(time,volt)
```

```python
plt.figure(figsize=(14,10), dpi=150)
plt.xlabel("Temperatura (ºC)")
plt.ylabel("Resistencia $\Omega$")
plt.grid('on')
plt.plot(time,volt)
plt.ylim([0,60000])
plt.xlim([-10,60])
```

```python
plt.figure(figsize=(14,10), dpi=150)
plt.subplot(311)
plt.grid('on')
plt.plot(time,volt)
plt.ylim([50000,60000])
plt.xlim([-10,-8])
plt.subplot(312)
plt.grid('on')
plt.plot(time,volt)
plt.ylim([10000,16000])
plt.xlim([15,20])
plt.subplot(313)
plt.grid('on')
plt.plot(time,volt)
plt.ylim([3000,4000])
plt.xlim([48,52])
```

```python
import sympy as sym
```

```python
Rx, R1, Vi, C, S = sym.symbols('Rx R1 Vi C, S')
```

```python
Rp=(Rx*(1/(S*C)))/(Rx+(1/(S*C)))
```

```python
print(sym.simplify(Rp))
```

```python
Vx=(Rp*Vi)/(Rp+R1)
print(Vx)
```

```python
print(sym.simplify(Vx))
```

```python
sym.simplify((Rx*Vi)/((R1*C*Rx*S)+R1+Rx))
```

```python
sym.cancel((Rx*Vi)/((R1*C*Rx*S)+R1+Rx))
```

```python
sym.simplify(Vx.subs([(R1,10000),(Rx,10000),(Vi,5)]))
```
