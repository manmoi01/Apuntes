##Introducción

IPython es un herramienta que permite el desarrollo interactivo de comandos de python, a diferencia de ejectuar un script de forma estructurada ***Top-Down***, permite la ejecución de los comandos en elemntos llamados celdas *cells*. De está forma cada celda se ejecuta de forma independiente, por lo que se puede ejecutar el código de la celda cuantas veces sea necesario para tener el resultado deseado, sin tener que modificar o ejecutar el resto de las celdas. Es importante señalar que las variables son globales y se conservan despues de su ejecución para las demas celdas, así por ejemplo si se modifica la celda 3, es necesario ejecutar de nuevo la celda 4 y posteriores para actualizar su valor.  Ipython ofrece dos formas de interctuar, la primera es en consola, en este modo se tiene un sistema similar al de python y se ejecuta en una consola de salida. La seguna y más interesante forma es la de notebook la cual utilizando el web server **tornado** se crea una sesión interactiva via websockets y JSON, permitiendo interactuar a través de un **navegador web moderno**.

Este tutorial se enfoca en el sistema notebook que permite el desarrollo de paginas donde se incluyen comandos de python, este sistema esta limitado a que no permite entradas del usuario como ocurren la version de consola.

````python
from pylab import *

figure(figsize=(10,8), dpi=150)
subplot(111)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plot(X, S, color="red", linewidth=2.5, linestyle="-",  label="sine")

ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xlim(X.min()*1.1, X.max()*1.1)
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ylim(C.min()*1.1,C.max()*1.1)
yticks([-1, +1],
       [r'$-1$', r'$+1$'])


legend(loc='upper left')

t = 2*np.pi/3
plot([t,t],[0,np.cos(t)],
     color ='blue',  linewidth=1.5, linestyle="--")
scatter([t,],[np.cos(t),], 50, color ='blue')
annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)),  xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=19,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plot([t,t],[0,np.sin(t)],
          color ='red',  linewidth=1.5, linestyle="--")
scatter([t,],[np.sin(t),], 50, color ='red')
annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)),  xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=19,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(15)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))


# savefig("../figures/exercice_10.png",dpi=72)
show()
````

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

n_angles = 36
n_radii = 8

# An array of radii
# Does not include radius r=0, this is to eliminate duplicate points
radii = np.linspace(0.125, 1.0, n_radii)

# An array of angles
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

# Repeat all angles for each radius
angles = np.repeat(angles[...,np.newaxis], n_radii, axis=1)

# Convert polar (radii, angles) coords to cartesian (x, y) coords
# (0, 0) is added here. There are no duplicate points in the (x, y) plane
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

# Pringle surface
z = np.sin(-x*y)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.2)

plt.show()
````
