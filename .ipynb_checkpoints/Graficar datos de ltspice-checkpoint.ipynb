##Graficar datos de espectros FFT
Es importante recordar que el archivo se tiene que guardar con la opcion de cordendas complejas cartesianas x+iy y no angular

```python
import numpy as np
import pylab
from scipy.signal import argrelextrema

filename = 'modulardoam[1].txt'

# Using the newer with construct to close the file automatically.
with open(filename) as f:
    data = f.readlines()

data2=[]
for i in data[1:len(data)]:
    data2.append(i.strip())
tam1=len(data2)
flag=0
if tam1 > 1:
    tam2=len(data2[0].split('\t'))
    datos=np.zeros((tam1,tam2),dtype=numpy.complex64)

for i in range(tam1):
    s=data2[i].split('\t')
    s3=np.array([],dtype=complex64)
    for k in s:
        s2=np.asarray(k.split(','),dtype=np.float32)
        if np.shape(s2) == (1,):
            s3=np.append(s3,s2)
        else:
            s3=np.append(s3,s2[0]+1j*s2[1])
    datos[i]=s3
s4=np.shape(datos)
matplotlib.rcParams.update({'font.size': 20,'text.usetex': True})
figure(figsize=(14,10), dpi=150)
xlim(5,2e4)
ylim(-180,-55)
for i in range(1,s4[1]):
    semilogx(np.real(datos[:,0]),20*np.log10(abs(datos[:,i])))
grid()
```

Valores para los sistemas con numeros complejos, valores asigandos a las partes numericas que corresponden a la calumna compleja

compleja -> obtenios

1 -> 1,2

2 -> 3,4

3 -> 5,6

4 -> 7,8

5 -> 9,10

```python
t=[]
i1=[]
i2=[]
for i in data2:
    s=i.split('\t')
    t.append(s[0])
    i1.append(s[1].split(','))
    i2.append(s[2].split(','))

i2n=np.asarray(i2,dtype=np.float32)   
i1n=np.asarray(i1,dtype=np.float32)   
t=np.asarray(t,dtype=np.float)  
```

```python
i1f=i1n[:,1]+1j*i1n[:,0]
i2f=i2n[:,1]+1j*i2n[:,0]
i1max=np.array([])
i2max=np.array([])
tmax=np.array([])
for i in range(0,len(t)):
    if 20*np.log10(np.abs(i1f[i])) > -120:
        i1max=np.append(i1max,i1f[i])
        i2max=np.append(i2max,i2f[i])
        tmax=np.append(tmax,t[i])
index=argrelextrema(np.abs(i1max), np.greater)
indexm=np.argmax(np.abs(i1max))
index2=argrelextrema(np.abs(i2max), np.greater)
indexm2=np.argmax(np.abs(i2max))
```

```python
matplotlib.rcParams.update({'font.size': 20,'text.usetex': True})
figure(figsize=(14,10), dpi=150)
xlim(5,1e5)
ylim(-180,-50)
xlabel('Frecuencia $\omega$')
ylabel('Amplitu (dB)')
semilogx(tmax[index],20*np.log10(np.abs(i1max[index])),'rx')
semilogx(tmax[index2],20*np.log10(np.abs(i2max[index2])),'ko')
semilogx(t,20*np.log10(np.abs(i1f)))
semilogx(t,20*np.log10(np.abs(i2f)))
annotate(str(tmax[indexm])+','+str(round(20*np.log10(np.abs(i1max[indexm])),2)), 
         xy=(tmax[indexm],20*np.log10(np.abs(i1max[indexm]))),
         xycoords='data',xytext=(-100, 20), textcoords='offset points',
         arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"),
         color = '#0000ee')
annotate('Primer maximo \n'+str(tmax[1])+','+str(round(20*np.log10(np.abs(i1max[1])),2)),
         xy=(tmax[1],20*np.log10(np.abs(i1max[1]))),  xycoords='data',xytext=(-70, 40),
         textcoords='offset points',arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"),
         color = '#0000ee')
annotate('Primer maximo \n'+str(tmax[indexm2])+','+str(round(20*np.log10(np.abs(i2max[indexm2])),2)), 
         xy=(tmax[indexm2],20*np.log10(np.abs(i2max[indexm2]))),
         xycoords='data',xytext=(50, 20), textcoords='offset points',
         arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"),
         color = '#009900')
grid()
savefig('images/espectroam.png')
```

##Grafica de sistemas con multiples cambios
Estos archivos se obtienen cuando se utiliza el comando .step param 

```python
filename = 'tanquelc2.txt'

# Using the newer with construct to close the file automatically.
with open(filename) as f:
    data2 = f.readlines()
k=0
inde=[]
for i in data2:
    if 'Step' in i:
        inde=append(inde,k)
    k +=1
inde=append(inde,len(data2))
k=len(inde)

leninde=[]
for i in range(0,k-1):
    leninde=append(leninde,inde[i+1]-inde[i])

print("Se encontraron {} pasos en la simulacion".format(k-1))
print(leninde)
```

```python
inter=6
datainter=[]
int(inde[inter-1].tolist()+1)
for i in range(int(inde[inter-1]+1),int(inde[inter]-1)) :
    datainter = append(datainter,data2[i].strip())

    
tam1=len(datainter)
tam2=len(datainter[1].split('\t'))

datainter2=np.zeros((tam1,tam2))
for i in range(tam1):
    s=datainter[i].split('\t')
    for k in range(tam2):
        datainter2[i,k]=np.asarray(s[k],dtype=np.float32)
plot(datainter2[:,0],datainter2[:,1:tam2])
```

#Curvas Caracteriticas NMOS
##$V_{DS}$ vs $I_{D}$

```python
#ioff()
filename = 'nmos_curva.txt'

# Using the newer with construct to close the file automatically.
with open(filename) as f:
    data2 = f.readlines()
k=0
inde=[]
for i in data2:
    if 'Step' in i:
        inde=append(inde,k)
    k +=1
inde=append(inde,len(data2))
k=len(inde)

leninde=[]
for i in range(0,k-1):
    leninde=append(leninde,inde[i+1]-inde[i])

print("Se encontraron {} pasos en la simulacion".format(k-1))
print(leninde)

matplotlib.rcParams.update({'font.size': 20,'text.usetex': True})
figure(figsize=(14,10), dpi=150)
print(data2[0].strip())

for inter in range(1,5):
    datainter=[]
    for i in range(int(inde[inter-1]+1),int(inde[inter])) :
        datainter = append(datainter,data2[i].strip())
           
    tam1=len(datainter)
    tam2=len(datainter[1].split('\t'))

    datainter2=np.zeros((tam1,tam2))
    for i in range(tam1):
        s=datainter[i].split('\t')
        for k in range(tam2):
            datainter2[i,k]=np.asarray(s[k],dtype=np.float32)
    plot(datainter2[:,0],datainter2[:,1])
xlabel('V_{DS}')
ylabel('$I_{d}$')
legend(['$V_{GS}=2$','$V_{GS}=2.8$','$V_{GS}=4$','$V_{GS}=5$'],loc=2)
grid()
xlim(0,5)
ylim(0,0.00026)
title('Transistor NMOS del SPICE')
savefig('nmos_ideal.png')
```
